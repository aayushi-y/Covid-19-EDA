```python
#creating visualization for world dataset 
worldometer_df_sum = worldometer_df.groupby(by = 'Continent').sum()
print(worldometer_df_sum)
```

```python
import plotly.express as px 
x = ['Africa', 'Asia', 'Australia/Oceania', 'Europe', 'North America', 'South America']
fig1 = px.bar(worldometer_df_sum,x = x, y = 'TotalCases', color = x)
fig1.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,8))
N = 6
ind = np.arange(N) 
width = 0.25

bar1 = plt.bar(ind,worldometer_df_sum['TotalDeaths'] , width, color = 'red')
  
bar2 = plt.bar(ind+width, worldometer_df_sum['TotalRecovered'], width, color='green')
  
bar3 = plt.bar(ind+width*2,worldometer_df_sum['ActiveCases'], width, color = 'cyan')
  
plt.xlabel("Continent")
plt.ylabel('Count')
plt.title("Continent wise data")
  
plt.xticks(ind+width,['Africa', 'Asia', 'Australia/Oceania', 'Europe', 'North America', 'South America'])
plt.legend( (bar1, bar2, bar3), ('Total Deaths', 'Total Recovered', 'Active cases') )
plt.show()
```

Now,plotting the GDP of various countries to see if we find any interersting pattern between them. 

```python
gdp_df = pd.read_csv("/kaggle/input/gdp-of-all-countries19602020/gdp_1960_2020.csv")
gdp_df = gdp_df.copy().loc[gdp_df['year'] == 2020]
gdp_df.head(10)
```

```python
gdp_df.rename(columns = {'country':'Country/Region'}, inplace = True)
gdp_df.columns
```

```python
gdp_df.shape
```

```python
top_10_gdp_2020 = gdp_df.sort_values('gdp', ascending = False).head(10)
top_10_gdp_2020
```

```python
#top_10_gdp_2020 = top_10_gdp_2020.merge(worldometer_df, on = 'Country/Region')
#top_10_gdp_2020

labels = 'Africa','America', 'Asia','Europe', 'Australia/Oceania' 

fig = plt.figure()
plt.pie(gdp_df.groupby(by = 'state')['gdp_percent'].sum(), labels=labels, autopct='%1.1f%%')
plt.title('GDP distribution 2020')
plt.show()
```

```python
fig = plt.figure()
plt.pie(worldometer_df.groupby(by = 'Continent')['Population'].sum(), labels=x, autopct='%1.1f%%')
plt.title('Population distribution 2020')
plt.show()
```

### Following interseting observations can be made :
**1. United States has accounted for the maximum number of deaths across the world. This is significant observation as USA has the highest GDP across the globe.**

**2. Asia, on the other hand, accounting for the majority portion of the world population, ranking second in GDP contribution fared best when it came to the mortality rate due to COVID 19.**
